```python
import pandas as pd
```

```python
#Leitura dos arquivos
df1 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/datasets/Aracaju.xlsx")
df2 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/datasets/Fortaleza.xlsx")
df3 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/datasets/Natal.xlsx")
df4 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/datasets/Recife.xlsx")
df5 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/datasets/Salvador.xlsx")
```

```python
#exibindo as linhas do arquivo
df1.head()
```

```python
#juntando todos os arquivos
df = pd.concat([df1, df2, df3, df4, df5])
```

```python
#exibindo as 5 primeiras linhas do arquivo
df.head()
```

```python
#exibindo as 5 últimas linhas do arquivo
df.tail()
```

```python
#Visualizar uma amostra do conteúdo do arquivo
df.sample(5)
```

```python
#Verificando o tipo de dados de cada coluna
df.dtypes
```

```python
#Alterar tipo de dado
df["LojaID"] = df["LojaID"].astype("object")
```

```python
#Verificando o tipo de dados de cada coluna
df.dtypes
```

**Tratando Valores Faltantes**


```python
#Consultando linhas com valores faltantes
df.isnull().sum()
```

```python
#Substituíndo os valores nulos pela média
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
```

```python
df["Vendas"].mean()
```

```python
#Consultando linhas com valores faltantes
df.isnull().sum()
```

```python
#Substituíndo os valores nulos por zero
df["Vendas"].fillna(0, inplace=True)
```

```python
#Apagando linhas com valores nulos
df.dropna(inplace=True)
```

```python
#Apagando as linhas com valores nulos com base apenas em uma coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
#Remover linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
```

Criando novas Colunas



```python
#Criando coluna de receita(Vendas * Qtde)
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
#Retornando a maior receita
df["Receita"].max()
```

```python
#Retornando a menor receita
df["Receita"].min()
```

```python
#nlargest, retorna a quantidade de linhas (definir na linha de comando) com as maiores receitas
df.nlargest(5, "Receita")
```

```python
#nsamllest, retorna a quantidade de linhas (definir na linha de comando) com as menores receitas
df.nsmallest(5, "Receita")
```

```python
#Agrupamento por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#Ordenando o conjunto de dados
df.sort_values("Receita", ascending=False).head(10)
```
